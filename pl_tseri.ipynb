```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('data2020/sha_bjfs.neu', comments='#')
#print(data)

t = data[:,0]
n = data[:,1]*1000
e = data[:,2]*1000
u = data[:,3]*1000

plt.figure()    # 定义一个图像窗口
plt.plot(t,n,color="blue", linewidth=1.0, linestyle="-", label="N") # 绘制曲线 n
plt.plot(t,e,color="green", linewidth=1.0, linestyle="-", label="E") # 绘制曲线 e
plt.plot(t,u,color="red", linewidth=1.0, linestyle="--", label="U") # 绘制曲线 u
plt.xlabel('Year')
plt.ylabel('mm')
plt.title('BJFS')
plt.legend(loc="lower left")
plt.savefig('./BJFS.pdf')
plt.show()

```
