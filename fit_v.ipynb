```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('data2020/sha_bjfs.neu', comments='#')
#print(data)

t = data[:,0]
n = data[:,1]*1000
e = data[:,2]*1000
u = data[:,3]*1000

# fit data with polynomial
# y = a*x + b;
# s = v*t + offset;
#f1 = np.polyfit(t, e, 1)  # 用1次多项式拟合，可改变多项式阶数；
[a,b] = np.polyfit(t, e, 1) 
print([a,b])
#print('f1 is :\n',a)
p1 = np.poly1d([a,b])  #得到多项式系数，按照阶数从高到低排列
print('p1 is :\n',p1)
#也可使用yvals=np.polyval(f1, x)
yvals = p1(t)  #拟合y值
#print('yvals is :\n',yvals)
#绘图
plot1 = plt.plot(t, e, '-',label='original values')
plot2 = plt.plot(t, yvals, 'r',label='polyfit values')
#plt.text(2012,0,f1[0])
plt.text(2012,0,a)
plt.xlabel('year')
plt.ylabel('e/mm')
plt.legend(loc=0) #指定legend的位置右下角
plt.title('polyfitting')
plt.show()
```
